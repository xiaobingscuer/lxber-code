# 随机(random)与随机漫步（random-walk）的区别

####  区别：   
> 绝对随机是无任何规律的，混乱的，无序的   
> 随机漫步具有随机性，并且也具有向有序趋近的性质。这种向有序趋近的的特征是由于分形造成的混沌系统的结果。  

## 绝对随机的几何

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(1000)
```

```python

plt.plot(data)
plt.show()
```

## 随机漫步的几何

```python
plt.plot(data.cumsum())
plt.show()
```
